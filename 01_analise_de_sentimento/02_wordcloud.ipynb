```python
import pandas as pd

resenha = pd.read_csv('dados/imdb-reviews-pt-br.csv')
resenha.head()
```

```python
%matplotlib inline

from wordcloud import WordCloud
```

```python
todas_palavras = ' '.join([texto for texto in resenha.text_pt])
```

```python
nuvem_palavras = WordCloud(
    width=800, height=500, 
    max_font_size=110,
    collocations=False).generate(todas_palavras)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
def nuvem_palavra_neg(texto, coluna_texto):
    texto_negativo = texto[texto['sentiment'] == 'neg']

    todas_palavras = ' '.join([texto for texto in texto_negativo[coluna_texto]])

    nuvem_palavras = WordCloud(
        width=800, height=500,
        max_font_size=110,
        collocations=False).generate(todas_palavras)

    plt.figure(figsize=(10, 7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
def nuvem_palavra_pos(texto, coluna_texto):
    texto_negativo = texto[texto['sentiment'] == 'neg']

    todas_palavras = ' '.join([texto for texto in texto_negativo[coluna_texto]])

    nuvem_palavras = WordCloud(
        width=800, height=500,
        max_font_size=110,
        collocations=False).generate(todas_palavras)

    plt.figure(figsize=(10, 7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
nuvem_palavra_neg(resenha, 'text_pt')
```

```python
nuvem_palavra_pos(resenha, 'text_pt')
```
